```python
# change all print statements to return.

import math

firstNum = 0
nextNum = 0
total = 0
operator = ''


def add(num1, num2): # addition function
    return num1 + num2

def subtract(num1, num2): # subtraction function
    return num1 - num2

def multiply(num1, num2): # multiplication function
    return num1 * num2

def divide(num1, num2): # division function
    if num2 == 0: # checks for divide by 0 error
        return "Cannot divide by zero"
    return num1 / num2

def reset(): # reset function
    global total, firstNum, nextNum, operator
    total = 0
    firstNum = 0
    nextNum = 0
    operator = ''

def main(): 
    reset()

    while True: # loop for checking for valid input
        firstNum = input("Enter a number: ")
        try:
            firstNum = float(firstNum)
            break
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

    total = firstNum
    print("Current total:", total)

    while True: # loop for recalculation
        operator = input("Enter the operation you want to perform (+, -, x, /) (! to exit): ")
        if operator == '!':
            break
        
        while True: # checking for input validity of second number
            nextNum = input("Enter a number: ")
            try:
                nextNum = float(nextNum)
                break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")

        
        if operator == '+':
            total = add(total, nextNum)
        elif operator == '-':
            total = subtract(total, nextNum)
        elif operator == 'x':
            total = multiply(total, nextNum)
        elif operator == '/':
            result = divide(total, nextNum)
            if isinstance(result, str): 
                print(result)
                continue
            total = result
        else:
            print("Invalid input, try again")
            continue
        
        print("Current total:", total)

if __name__ == "__main__":
    main()
```
